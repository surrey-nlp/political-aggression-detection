```python
import numpy as np
import pandas as pd
from sklearn.metrics import cohen_kappa_score
```

```python
elections_14_af = pd.read_csv('merged_14_after.csv')
elections_19_af = pd.read_csv('merged_19_after.csv')
```

```python
elections_14_af.head(5)
```

```python
labels_Dnyaneshwar = elections_14_af['Dnyaneshwar_Label'].to_numpy()
print(labels_Dnyaneshwar)
```

```python
labels_Nazia = elections_14_af['Nazia_Label'].to_numpy()
```

```python
election_14_after_kappa = cohen_kappa_score(labels_Dnyaneshwar, labels_Nazia)
print(election_14_after_kappa)
```

```python
labels_Nazia_2 = elections_19_af['Nazia_Label'].to_numpy()
labels_Dnyaneshwar_2 = elections_19_af['Dnyaneshwar_Label'].to_numpy()
```

```python
election_19_after_kappa = cohen_kappa_score(labels_Dnyaneshwar_2, labels_Nazia_2)
print(election_19_after_kappa)
```
